# Q1. Check and remove missing data
### Q1.1 Write a Python code snippet that checks for missing values in each column of the dataset. If any, only display the names of the columns with missing values and their corresponding number of missing values. Print the length of the data before removing the missing data.

```python
import pandas as pd

# Read the dataset
df = pd.read_csv('Files/Real_Estate_Sales_2001-2020_GL.csv')
# Check for missing values in each column
missing_values = df.isnull().sum()

# Filter columns with missing values
columns_with_missing_values = missing_values[missing_values > 0]

# Display column names and number of missing values
if len(columns_with_missing_values) > 0:
    print("Columns with missing values:")
    for column in columns_with_missing_values.index:
        print(f"{column}: {columns_with_missing_values[column]} missing values")
else:
    print("No missing values found in the dataset.")
```

```python
import pandas as pd

# read the dataset into a pandas dataframe
df = pd.read_csv('Files/Real_Estate_Sales_2001-2020_GL.csv')

# print the length of the original data set
print(f"Length of data before removing missing values:", (len(df)))

# check for missing values in each column
missing_values = df.isnull().sum()

# iterate over columns with missing values and remove them
for column in df.columns:
    if missing_values[column] > 0:
        print(f"Column '{column}' has {missing_values[column]} missing values.")
        df.dropna(subset=[column], inplace=True)

# print the length of the resulting cleansed data set
print(f"Length of data after removing missing values: {len(df)}")
```

```python
import pandas as pd

# Read the dataset
df = pd.read_csv('Files/Real_Estate_Sales_2001-2020_GL.csv')

# Specify the column name for which you want to compute summary statistics
column_name = 'OPM remarks'

# Filter out non-numeric values
filtered_column = pd.to_numeric(df[column_name], errors='coerce')

# Compute summary statistics for the filtered column
count = filtered_column.count()
mean = filtered_column.mean()
std_dev = filtered_column.std()
minimum = filtered_column.min()
maximum = filtered_column.max()

# Print the summary statistics
print(f"Count: {count}")
print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
print(f"Minimum: {minimum}")
print(f"Maximum: {maximum}")
```
