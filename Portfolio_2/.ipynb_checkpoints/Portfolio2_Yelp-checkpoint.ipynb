### Q1.3 We want to remove categories that appear fewer than 200 times. Write a Python code snippet to perform this operation. At last, print the number of resulting unique business_categories.

```python
# your code and solutions
```

```python
import pandas as pd

# Load the data into a pandas DataFrame
data = pd.read_csv("Files/Yelp_Portfolio2_Input.csv")

# Filter the categories that appear fewer than 200 times
category_counts = data['business_categories'].str.split('; ').explode().value_counts()
valid_categories = category_counts[category_counts >= 200].index.tolist()

# Filter the DataFrame to include only rows with the remaining categories
data = data[data['business_categories'].str.contains('|'.join(valid_categories))]
```

```python
# Print the number of resulting unique business_categories
print("Number of resulting unique business_categories:", len(data['business_categories'].str.split('; ').explode().unique()))
```

```python
import pandas as pd
import re

# Load the data into a pandas DataFrame
data = pd.read_csv("Files/Yelp_Portfolio2_Input.csv")

# Filter the categories that appear fewer than 200 times
category_counts = data['business_categories'].str.split('; ').explode().value_counts()
valid_categories = category_counts[category_counts >= 200].index.tolist()

# Escape any special characters in the category names
valid_categories_escaped = [re.escape(cat) for cat in valid_categories]

# Filter the DataFrame to include only rows with the remaining categories
data = data[data['business_categories'].str.contains('|'.join(valid_categories_escaped))]

# Print the number of resulting unique business_categories
print("Number of resulting unique business_categories:", len(data['business_categories'].str.split('; ').explode().unique()))
```

```python
import pandas as pd

# Load the data into a pandas DataFrame
data = pd.read_csv("Files/Yelp_Portfolio2_Input.csv")

# Count the number of occurrences for each business category
category_counts = data['business_categories'].str.split('; ').explode().value_counts()

# Filter the categories that appear fewer than 200 times
valid_categories = category_counts[category_counts >= 200].index.tolist()

# Filter the DataFrame to include only rows with the remaining categories
data = data[data['business_categories'].str.split('; ').apply(set).apply(lambda x: set(valid_categories).issuperset(x))]

# Print the number of resulting unique business_categories
print("Number of resulting unique business_categories:", len(data['business_categories'].str.split('; ').explode().unique()))
```

# Q2. Outliers

### Q2.1 Remove any rows from the sample_data DataFrame (output from Q1.3) where the `useful` column has a value of 6 or more; Then, create a single figure with two subplots, one showing the boxplot of `useful` column of the original data and the other showing the boxplot of the data with outliers removed. 

At last, __print the length of the data__.

```python
# your code and solutions
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas DataFrame
data = pd.read_csv("Files/Yelp_Portfolio2_Input.csv")

# Remove any rows where the useful column has a value of 6 or more
data = data[data['useful'] < 6]

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Add a title to the figure
fig.suptitle("Boxplots of 'useful' column")

# Create a boxplot of the 'useful' column of the original data
ax1.boxplot(data['useful'])
ax1.set_title("Original data")

# Create a boxplot of the 'useful' column of the filtered data
ax2.boxplot(data[data['useful'] < 6]['useful'])
ax2.set_title("Data with outliers removed")

# Add x-axis labels to both subplots
ax1.set_xlabel("Original data")
ax2.set_xlabel("Data with outliers removed")

# Display the figure
plt.show()
```

### Q2.2 In general, we might remove some inactive users from a dataset (output from Q2.1), for example, users who rate businesses less than 3 times. However, in this case, we are doing the opposite and removing extremely active users who rate businesses more than 30 times (>30). Again, create a single figure with two subplots, one showing the boxplot of the count of business rating given by each user and the other showing the boxplot of the data with outliers removed. 

At last, __print the length of the data__.

```python
# your code and solutions
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv('Files/Yelp_Portfolio2_Input.csv')
```

```python
# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))

# subplot 1: boxplot of the count of business rating given by each user
ax1.boxplot(df.groupby('user_id')['business_id'].count())
ax1.set_title('Original Data')

# subplot 2: boxplot of the data with outliers removed
active_users = df['user_id'].value_counts()
active_users = active_users[active_users <= 30].index
df_filtered = df[df['user_id'].isin(active_users)]
ax2.boxplot(df_filtered.groupby('user_id')['business_id'].count())
ax2.set_title('Data with Outliers Removed')

# display the figure
plt.show()
```

```python
print('Length of filtered data:', len(df_filtered))
```

# Q3. Training a LinearRegression model on the data output from Q2.2.

### Q3.1 How to build a linear regression model on a subset of the data, using the `useful` column as the input variable and the `stars` column as the output variable. Split the data into a training set and a test set, with 70% of the data used for training. The random_state is set to 42.

```python
# your code and solutions
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv('Files/Yelp_Portfolio2_Input.csv')
```

```python
subset_df = df[['useful', 'stars']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(subset_df['useful'], subset_df['stars'], test_size=0.3, random_state=42)
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
y_pred = lr_model.predict(X_test.values.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE:', mse)
print('R-squared:', r2)
```
